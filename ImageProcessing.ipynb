```python
#Blending images is done through the addWeighted function that uses both images and combines them
#To blend images we use a simple formula: new_pixel = (alpha)*pixel_1 + (beta)*pixel_2+ gamma
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('Earth-8-2880x1800.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('images.jpeg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img1.shape
plt.imshow(img1)
```

```python
img2.shape
plt.imshow(img2)
```

```python
img1 = cv2.resize(img1, (1200, 1200))
img2 = cv2.resize(img2, (1200, 1200))
```

```python
blended = cv2.addWeighted(src1=img1, alpha=0.7, src2=img2, beta=0.3, gamma=0) #addWeighted works when images 
#are of same size 
```

```python
plt.imshow(blended)
```
